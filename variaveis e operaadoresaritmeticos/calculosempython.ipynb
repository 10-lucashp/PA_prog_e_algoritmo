Calculo de média

```python
## Digite as notas de Ana na faculdade
nota1 = float(input("Digite a nota 1:"))
nota2 = float(input("Digite a nota 2:"))
```

```python
## Calcule a média de Ana
media = (nota1 + nota2)/2
```

```python
##Mostra a média
print(f"A média de Ana no bimestre foi:{media} ")
```

Conversão de Dólar pro Real

```python
##Recebe a cotação do dollar e o valor em reais
dolarcotacao = float(input("Coloque a cotação atual do dolar: "))
valoremreais = float(input("Coloque o valor em reais: "))
```

```python
##Calculo de conversão
dolarvalor = valoremreais / dolarcotacao
```

```python
##Resultado
print(f"O valor em dolares é: {dolarvalor:.2f} ")
```

Cálculo de comissão

```python
##Cálculo de comissão
precounitario = float(input("Coloque o preço unitário do produto: "))
quantidadedevendas = float(input("Coloque a quantidade vendida: "))
```

```python
##Calculo
valortotalvenda = precounitario * quantidadedevendas
comissao = valortotalvenda * 0.05
```

```python
##Resultado
print(f"Valor total da venda: R$`{valortotalvenda:.2f} ")
print(f"Comissão do vendedor: R$ {comissao:.2f}")
```
